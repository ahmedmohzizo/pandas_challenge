```python
# Import Dependencies
import pandas as pd
```

```python
school_data_to_load = "Resources/schools_complete.csv"
school_data_df = pd.read_csv(school_data_to_load)
school_data_df.head()
```

```python
student_data_to_load = "Resources/students_complete.csv"
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```

```python
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
#Calculate the total number of schools
total_school_name = school_data_complete_df["school_name"].value_counts ()
total_school_name
```

```python
#Calculate the total budget
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
total_school_sum = total_school_name.value_counts()
total_school_sum
```

```python
total_school = total_school_sum.sum()
total_school
```

```python
#Calculate the total number of students
total_students_sum = school_data_complete_df["student_name"].value_counts()
total_students_sum
```

```python
total_students = total_students_sum.sum()
total_students
```

```python
average_math = school_data_complete_df["math_score"].mean()
average_math
```

```python
#Calculate the average reading score
average_reading = school_data_complete_df["reading_score"].mean()
average_reading
```

```python
#Calculate the percentage of students with a passing math score (70 or greater
math_sum = school_data_complete_df["math_score"].value_counts().sum()
math_sum
```

```python
math_score = school_data_complete_df["math_score"]>= 70
math_score = (math_score).mean()*100
math_score
```

```python
math_score_pass = school_data_complete_df["math_score"]>= 70
math_score_pass
```

```python
#Passing Reading
Passing_Reading = school_data_complete_df["reading_score"]>= 70
Passing_Reading = (Passing_Reading).mean()*100
Passing_Reading
```

```python
Passing_Reading_pass = school_data_complete_df["reading_score"]>= 70
Passing_Reading_pass
```

```python
 #Overall Passing (The percentage of students that passed math and reading.
overall_passing = ( math_score + Passing_Reading )/2.4
overall_passing
```

```python
#give the displayed data cleaner formatting
district_results = [{"total_school":total_school, 
            "Total Students": total_students, 
            "Total Budget": total_budget, 
            "average_math": round(average_math,2), 
            "average_reading_score": round(average_reading,2), 
           "percent_passing_math": round(math_score,2),
           "percent_passing_reading": round(Passing_Reading,2),
            "overall_passing": round(overall_passing,2)}]
district_summary_table = pd.DataFrame(district_results)

#Formatting
district_summary_table["percent_passing_math"] = district_summary_table["percent_passing_math"].map("{:,.2f}%".format)
district_summary_table["percent_passing_reading"] = district_summary_table["percent_passing_reading"].map("{:,.2f}%".format)
district_summary_table["overall_passing"] = district_summary_table["overall_passing"].map("{:,.2f}%".format)
district_summary_table["Total Budget"] = district_summary_table["Total Budget"].map("${:,.2f}".format)
district_summary_table["Total Students"] = district_summary_table["Total Students"].map("{:,}".format)

#Display
district_summary_table
```

```python
#Create an overview table that summarizes key metrics about each school, including:

school_data_complete_df ["passing_math"] = math_score_pass
school_data_complete_df ["passing_reading"] = Passing_Reading_pass
school_data_complete_df
```

```python
school_group = school_data_complete_df.groupby(["school_name"]).mean()
school_group["Per Student Budget"] = school_group["budget"]/school_group["size"]
school_group["percent_passing_math"] = round(school_group["passing_math"]*100,2)
school_group["percent_passing_reading"] = round(school_group["passing_reading"]*100,2)
school_group["overall_passing"] = round(((school_group["passing_math"] + school_group["passing_reading"])/2)*100,3)

#Merge with school_data to collect information about the type, size and budget
school_data_summary = pd.merge(school_group,school_data_complete_df, how="left", on=["school_name", "school_name"])
del school_data_summary['size_y']
del school_data_summary['budget_y']
del school_data_summary['School ID_x']
school_data_summary
```

```python
#Create a dataframe to store the results
school_summary_dataframe = pd.DataFrame({"School Name":  school_data_summary["school_name"],
                                "School Type": school_data_summary["type"],
                               "Total Students":school_data_summary["size_x"],
                               "Total School Budget": school_data_summary["budget_x"],
                               "Per Student Budget":school_data_summary["Per Student Budget"], 
                               "Average Math Score":round(school_data_summary["math_score_x"],2),
                               "Average Reading Score":round(school_data_summary["reading_score_x"],2), 
                               "% Passing Math": school_data_summary["percent_passing_math"],
                               "% Passing Reading": school_data_summary["percent_passing_reading"],
                               "% Overall Passing Rate": school_data_summary["overall_passing"]})
school_summary_dataframe["Total Students"] = school_summary_dataframe["Total Students"].map("{:,.0f}".format)
school_summary_dataframe["Total School Budget"] = school_summary_dataframe["Total School Budget"].map("${:,.2f}".format)
school_summary_dataframe["Per Student Budget"] = school_summary_dataframe["Per Student Budget"].map("${:,.2f}".format)

school_summary_dataframe
```

```python
top_five_schools = school_summary_dataframe.sort_values(["% Overall Passing Rate"], ascending=False)
top_five_schools.head()
```

```python
#Sort and display the five worst-performing schools by % overall passing
bottom_five_schools = school_summary_dataframe.sort_values(["% Overall Passing Rate"], ascending=True)
bottom_five_schools.head()
```

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"] 
school_summary_dataframe["Spending Ranges (Per Student)"] = pd.cut(school_data_summary["Per Student Budget"], spending_bins, labels=group_names)
school_spending_grouped = school_summary_dataframe.groupby("Spending Ranges (Per Student)").mean()

school_spending_grouped["Average Math Score"] = school_spending_grouped["Average Math Score"].map("{:,.2f}".format)
school_spending_grouped["Average Reading Score"] = school_spending_grouped["Average Reading Score"].map("{:,.2f}".format)
school_spending_grouped["% Passing Math"] = school_spending_grouped["% Passing Math"].map("{:,.2f}".format)
school_spending_grouped["% Passing Reading"] = school_spending_grouped["% Passing Reading"].map("{:,.2f}".format)
school_spending_grouped["% Overall Passing Rate"] = school_spending_grouped["% Overall Passing Rate"].map("{:,.2f}".format)
school_spending_grouped
```
